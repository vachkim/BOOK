### 1. Matplotlib 다운
- Matplotlib 패키지를 다운받는다

````
import matplotlib
import matplotlib.pyplot as plt  
````

### 2. 기본그래프 그리기

#### 1) Plot을 활용한 기본 그래프
- Plot은 직선 또는 꺾은선 그래프를 그리는 기본 함수이다
- 별도로 입력하지 않으면 입력값은 자동으로 Y축 값으로 인식된다

````
plt.plot([10,20,30,40])
plt.show()
````

- 다음과 같이 순차적으로 X,Y값을 지정해줄 수도 있다

````
plt.plot([1,2,3,4],[12,43,25,15])
plt.show()
````

#### 2) 그래프에 옵션 추가
- Matplotlib에서는 간단한 방식으로 제목, 범례 등의 옵션을 추가할 수 있다
- 제목을 추가하기 위해서는 Title옵션을 사용한다

````
plt.title('Plotting')
plt.plot([10,20,30,40])
plt.show()
````

- 데이터 구분을 위해 범례를 추가할 때는 각 데이터셋에 Label을 통해 이름을 부여하고, Legend옵션을 사용한다

````
plt.plot([10,20,30,40],label='asc')
plt.plot([40,30,20,10],label='desc')
plt.legend()
plt.show()
````

- 범례의 위치는 자동으로 지정되지만 loc를 활용하면 따로 위치값을 줄 수도 있다


````
plt.plot([10,20,30,40],label='asc')
plt.plot([40,30,20,10],label='desc')
plt.legend(loc=1)
plt.show()
````

- 그래프의 색상 변경을 위해서는 color옵션을 사용한다

````
plt.plot([10,20,30,40],color='skyblue',label='skyblue')
plt.plot([40,30,20,10],color='pink',label='pink')
plt.legend()
plt.show()
````

- 선 모양을 바꿀 때에는 linestyle을 사용한다

````
plt.plot([10,20,30,40],linestyle = '--', label='dash')
plt.plot([40,30,20,10],linestyle = ':',label='dotted')
plt.legend()
plt.show()
````

- 그래프의 선과 마커모양을 한번에 변경할 수도 있다

````
plt.plot([10,20,30,40],'r.', label='red circle') #앞의 r과 g는 각각 색깔, 뒤의 .과 ^는 각각 마커의 모양을 의미한다
plt.plot([40,30,20,10],'g^',label='green triangle')
plt.legend()
plt.show()
````

``` python 
import csv
f = open('기온데이터.csv',encoding='cp949')

result = [] #최고기온을 담을 빈 리스트를 만든다

plt.show()
```

#### 3) 조건에 맞는 데이터 시각화하기
- 특정 년도나 특정 날짜의 데이터를 모아 시각화해보자
- 날짜에 대한 정보를 추출하는 법은 다음과 같다

````
date = weather['날짜']
````

````
info = [] #날짜정보를 담을 빈 리스트를 만든다
for item in date:
  i = item.split('-') #위에서 보면 날짜는 object이기 때문에 '-'를 기준으로 split해주면 연월일 정보를 추출할 수 있다
  info.append(i)
````

````
#연 정보
year = []
for item in info:
  year.append(item[0])
print(year[:10])

#월 정보 
month = []
for item in info:
  month.append(item[1])
print(month[:10])

#일 정보
day = []
for item in info:
  day.append(item[2])
print(day[:10])
````

- 위 과정을 응용해 6월의 최고기온을 추출해 시각화하면 다음과 같다

````
june = []
for i, con in enumerate(weather['날짜']):
  item = con.split('-')
  if item[1] == '06':
    june.append(weather.iloc[i,4])
print(june[:10])
````

- 위를 시각화하면 다음과 같이 나타난다

````
plt.grid()
plt.title('June')
plt.plot(june,color = 'm'); 
````

- 앞서 본 사례와 마찬가지로 책은 리스트 형태로 데이터를 처리한다


- 최고기온과 최저기온을 함께 나타내려면 다음과 같이 진행하면 된다

````
high = []
low = []
for i, con in enumerate(weather['날짜']):
  item = con.split('-')
  if item[1] == '06':
    high.append(weather.iloc[i,4])
    low.append(weather.iloc[i,3])
plt.grid()
plt.title('June')
plt.plot(high,color = 'r')
plt.plot(low,color='b');
````

## 4. 다양한 시각화 방법
- 지금까지 살펴본 plt.plot외에 다양한 시각화 방법을 알아보자

### 1) 히스토그램
- 히스토그램은 자료의 분포를 막대 그래프 형식으로 나타낸 것이다

````
import matplotlib.pyplot as plt
plt.hist([1,1,2,3,4,5,6,6,7,8,10])
plt.show()
````

````
import random
dice = []
for i in range(5):
  dice.append(random.randint(1,6)) #1과 6사이의 정수를 5번 뽑아 dice에 입력한다
print(dice)
plt.hist(dice)
plt.show()
````

- 3이 2번, 4가 1번, 6이 2번 뽑혔으므로 히스토그램으로 나타내면 위와 같이 그려진다

### 2) 히스토그램 실습
- 기온데이터를 히스토그램으로 나타내보자
- Matplotlib은 기본적으로 리스트 형식을 입력값으로 받기 때문에 칼럼을 리스트로 변환하는 작업이 선행되어야한다

````
high = list(weather['최고기온(℃)']) #최고기온 데이터를 리스트형태로 저장한다
plt.hist(high, bins=100) # bins는 막대의 갯수를 의미한다, 즉 데이터셋을 얼마나 세밀하게 표현하는지를 정하는 기준이다
plt.show;
````

- 조건에 맞는 데이터를 히스토그램으로 그리려면 
- 앞서 plt.plot을 그릴 때와 같은 방식으로 조건을 부여하되, 함수식을 plt.hist로만 바꿔주면 된다

````
june = [] #6월달의 최고기온을 리스트형태로 저장
for i, con in enumerate(weather['날짜']):
  item = con.split('-')
  if item[1] == '06':
    june.append(weather.iloc[i,4])
plt.hist(june,bins=50);
````

### 3) 상자그림
- 상자그림은 데이터에서 최댓값, 최솟값, 4분위값을 표현하는 도표이다
- 데이터의 분포를 쉽게 확인할 수 있다는 장점이 있다

````
result = []
for i in range(13):
  result.append(random.randint(1,1000))
print(sorted(result)) #result값을 순차적으로 제시
plt.boxplot(result)
plt.show()
````
